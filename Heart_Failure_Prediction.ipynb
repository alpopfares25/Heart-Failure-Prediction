```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns #for visualization

#for classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import Perceptron
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier


#for feature selection
from sklearn.feature_selection import SelectPercentile
from sklearn.feature_selection import chi2,f_classif
from sklearn.feature_selection import GenericUnivariateSelect
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import VarianceThreshold

#for Normalization
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('/content/drive/MyDrive/heart_failure.csv')
data.head()
```

```python
#split data
y=data.DEATH_EVENT.values
x=data.drop("DEATH_EVENT",axis=1)
```

```python
#check dublicate features exist or not
duplicate_features=[]
for i in range(0,len(data.columns)):
  col_1=data.columns[i]
  for col_2 in data.columns[i+1]:
    if data[col_1].equals(data[col_2]):
      duplicate_features.append(col_2)
len(duplicate_features)
```

#Feature Selection



# 1- Constant Quasi

> in this way I will pass threshold=0.01 which means if variance of all values in column is less than 0.01 remove this column or that mean remove feature that 99% of it's values are similar



```python
Quasi_filter=VarianceThreshold(threshold=0.01)
x_0=Quasi_filter.fit_transform(x)
x_0.shape  #there is no column has values with less than 0.01 variance
```

```python
#Classification
x_train, x_test, y_train, y_test = train_test_split(x_0, y, test_size=0.2, stratify=y)

#normalize the data
scaler=StandardScaler()
x_train_std=scaler.fit_transform(x_train)
x_test_std=scaler.transform(x_test)

#using logisticRegression

LR=LogisticRegression()
LR.fit(x_train_std,y_train)
ac_lr = accuracy_score(y_test,LR.predict(x_test_std))

#using Perceptron
PR=Perceptron()
PR.fit(x_train_std,y_train)
ac_pr = accuracy_score(y_test,PR.predict(x_test_std))

#using SVM
svc=SVC()
svc.fit(x_train_std,y_train)
ac_svc = accuracy_score(y_test,svc.predict(x_test_std))

#using Decision Tree
DT= DecisionTreeClassifier(max_depth=4)
DT.fit(x_train_std,y_train)
ac_dt = accuracy_score(y_test,DT.predict(x_test_std))

#using Random Forest
RF= RandomForestClassifier(n_estimators=10, max_depth=5)
RF.fit(x_train_std,y_train)
ac_rf = accuracy_score(y_test,RF.predict(x_test_std))

#using KNN
knn= KNeighborsClassifier(n_neighbors=4)
knn.fit(x_train_std,y_train)
ac_knn = accuracy_score(y_test,knn.predict(x_test_std))

models = pd.DataFrame({
    'Model': ['Support Vector Machines', 'KNN', 'Logistic Regression', 'Random Forest', 'Perceptron', 'Decision Tree'],
    'Score': [ac_svc, ac_knn, ac_lr, ac_rf,ac_pr,ac_dt,]})
models.sort_values(by='Score', ascending=False)

```

# 2- Selection Percntile
